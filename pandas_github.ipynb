Återkoppling: Alright, #11 is incorrect. "Regions" should concern every region. That is, we want all the data in them to be groupby'd. Then calculate the "Rate per 10000", as in, number of victimizations per 10000 population.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('assults.csv', skipinitialspace = True)
```

```python
df
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
df.columns
```

```python
regions = df[['Region_2013_label','Rate_per_10000_population']]
regions
```

```python
regions['Rate_per_10000_population'] = regions['Rate_per_10000_population'].str.replace(',', '.')
```

```python
regions['Rate_per_10000_population'] = regions['Rate_per_10000_population'].replace('-   ', np.nan).astype(float)
print(regions.dtypes)
```

```python
regions
```

```python
regions_grouped = regions.groupby('Region_2013_label').Rate_per_10000_population.mean().reset_index()
regions_grouped
```
